```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.datasets._base import get_data_home

mnist = fetch_openml('mnist_784', version=1, cache=True)
N, d = mnist.data.shape
```

```python
def vectorize(i):
    vec = np.zeros((10, 1))
    vec[i] = 1
    return vec
```

```python
X_all = mnist.data.to_numpy(dtype='int32')
Y_all = mnist.target.to_numpy(dtype='int32')
X_train, X_test, Y_train, Y_test = train_test_split(X_all, Y_all, test_size=10000)
```

```python
y_train = [vectorize(y) for y in Y_train]
```

```python
x_train = [np.reshape(x, (784, 1)) for x in X_train]
```

```python
training_data = [(x / 255.0, y) for x, y in zip(x_train, y_train)]
x_test = [np.reshape(x, (784, 1)) for x in X_test]
testing_data = [(x / 255.0, y) for x, y in zip(x_test, Y_test)]
```
